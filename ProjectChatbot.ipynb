```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer()
from scipy.spatial.distance import cosine

for ans in range(0,42,6):
  dtm= count_vectorizer.fit_transform([data['Answer'][ans],data['Responses'][ans]])
  #print(pd.DataFrame(data=dtm.toarray(), columns=count_vectorizer.get_feature_names()))
  pd.DataFrame(data=dtm.toarray(), columns=count_vectorizer.get_feature_names())
  #print(cosine(dtm[0].toarray(),dtm[1].toarray())) 
  data.at[ans,'similarity count']= (1-cosine(dtm[0].toarray(),dtm[1].toarray()))
  # data.at[ans,'similarity count']=sim
data.head()
```
